```python
x = [(1.4,1.3,2.9) ,
             (1.8,1.4,3.2) ,
             (1.3,1.2,2.9) ,
             (0.9,3.5,3.1) ,
             (1.5,2.1,3.3)]
y=(1.25, 1.78, 3.01)
```

```python
Edistance=[]
for i in range(len(x)):
    Edistance.append(math.sqrt(sum([(a - b) ** 2 for a, b in zip(x[i], y)])))
```

```python
dfeuc1 = pd.DataFrame(x, columns = ['A' , 'B', 'C'], index=['x1','x2','x3','x4','x5']) 
#df = pd.DataFrame({'A': [0, 4, 5, 6, 7, 7, 6,5]})
dfeuc1['Euclidian Distance for our point'] = np.array(Edistance)
```

```python
Mdistance=[]
for i in range(len(x)):
    Mdistance.append(abs(x[i][0]-y[0])+abs(x[i][1]-y[1])+abs(x[i][2]-y[2]))
```

```python
dfman1 = pd.DataFrame(x, columns = ['A' , 'B', 'C'], index=['x1','x2','x3','x4','x5']) 
#df = pd.DataFrame({'A': [0, 4, 5, 6, 7, 7, 6,5]})
dfman1['Manhatten Distance for our point'] = np.array(Mdistance)
```

```python
Mindistance=[]
for i in range(len(x)):
    Mindistance.append(distance.minkowski(x[i],y,3))
#print(Mindistance)   
```

```python
dfmin1 = pd.DataFrame(x, columns = ['A' , 'B', 'C'], index=['x1','x2','x3','x4','x5']) 
#df = pd.DataFrame({'A': [0, 4, 5, 6, 7, 7, 6,5]})
dfmin1['minkowski Distance for our point'] = np.array(Mindistance)
```

```python
p=float('inf')
supdistance=[]
for i in range(len(x)):
    supdistance.append(distance.minkowski(x[i],y,p))    
    
```

```python
dfsup1 = pd.DataFrame(x, columns = ['A' , 'B', 'C'], index=['x1','x2','x3','x4','x5']) 
dfsup1['Supreme Distance for our point'] = np.array(supdistance)
```

```python
cosdistance=[]
for i in range(len(x)):
    cosdistance.append(distance.cosine(x[i],y))    
    
```

```python
dfcos1 = pd.DataFrame(x, columns = ['A' , 'B', 'C'], index=['x1','x2','x3','x4','x5']) 
dfcos1['Euclidian Distance for our point']=dfeuc1['Euclidian Distance for our point']
dfcos1['Manhatten Distance for our point']=dfman1['Manhatten Distance for our point']
dfcos1['minkowski Distance for our point']=dfmin1['minkowski Distance for our point']
dfcos1['Supreme Distance for our point']=dfsup1['Supreme Distance for our point']
dfcos1['cosine similarity for our point'] = np.array(cosdistance)
dfcos1
```

<h2>Question4(e)</h2>

```python
dat4=dat3.copy()
arr = dat4.to_numpy()
```

```python
#function for taking initial kmediods randomly from the data
def initmed(arr):#This functions take the input distance matrix and assign data points to the initial cluster which are randomly selected and returns 3 lists containing the data values of every cluster
    seed(1)
    s = choice(len(arr), size=3, replace=False)
    clus=[]#This list contains the data about which data point belongs to which cluster
    clus1=[]#This list contains the data values of the points which belongs to which cluster1
    clus2=[]#This list contains the data values of the points which belongs to which cluster2
    clus3=[]#This list contains the data values of the points which belongs to which cluster3
    for i in range(len(arr)):
        l1=distance.minkowski(arr[i],arr[s[0]],2)#euclidian distance between the data point and the kmediod 1
        l2=distance.minkowski(arr[i],arr[s[1]],2)#euclidian distance between the data point and the kmediod 2
        l3=distance.minkowski(arr[i],arr[s[2]],2)#euclidian distance between the data point and the kmediod 3
        if((l1<l2)and(l1<l3)):
            clus.append(0)
            clus1.append(list(arr[i]))
        elif((l2<l1)and(l2<l3)):
            clus.append(1)
            clus2.append(arr[i])
        else:
            clus.append(2)
            clus3.append(arr[i])
    dat4['After rearranging initial']=clus 
    return clus1,clus2,clus3
```

```python
clus1,clus2,clus3=initmed(arr)
for j in itertools.count():
    clus=[]
    #print(len(clus1))
    sai1=pd.DataFrame(squareform(pdist(clus1)))
    #print(sai1)
    w1=np.argmin(sai1.sum(axis=0))
    #print(w1)
    sai2=pd.DataFrame(squareform(pdist(clus2)))
    w2=np.argmin(sai2.sum(axis=0))
    sai3=pd.DataFrame(squareform(pdist(clus3)))
    #print(agg3)
    w3=np.argmin(sai3.sum(axis=0))
    sear=list(dat4.iloc[:,-1])
    #print(w3)
    #print(sear)
    o=[i for i, n in enumerate(sear) if n == 0][w1]
    o1=[i for i, n in enumerate(sear) if n == 1][w2]
    o2=[i for i, n in enumerate(sear) if n == 2][w3]
    #print(o)
    s=[o,o1,o2]
    #print(s)
    for i in range(len(arr)):
        l1=distance.minkowski(arr[i],arr[s[0]],2)
        l2=distance.minkowski(arr[i],arr[s[1]],2)
        l3=distance.minkowski(arr[i],arr[s[2]],2)
        if((l1<l2)and(l1<l3)):
            clus.append(0)
            clus1.append(arr[i])
        elif((l2<l1)and(l2<l3)):
            clus.append(1)
            clus2.append(arr[i])
        else:
            clus.append(2)
            clus3.append(arr[i])
    #print(len(clus))        
    dat4['After rearranging {0}'.format(j)]=clus
    if(dat4.iloc[:,-1].equals(dat4.iloc[:,-2])):
        #print('yes')
        break
```

<h5>Question 5(a) By Labeleling the clusters by the Type of music</h5>

<h6>complete Linkage</h6>

```python
agg6=pd.DataFrame(squareform(pdist(x20)))
Z8 = linkage(agg6, 'complete')
post1=music_data['Type'].tolist()
fig = plt.figure(figsize=(10, 20))
dn = dendrogram(Z8,orientation='left',labels=post1)
plt.tick_params(axis='y', labelsize=10)
plt.tick_params(axis='x', labelsize=10)
plt.ylabel('Type')
plt.xlabel('height')
plt.title('complete linkage Dendogram for label Music Type ')
plt.show()
```
